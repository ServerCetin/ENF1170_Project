```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

df = pd.read_csv("data/LA_crime_data_2023.csv")
df.columns
```

```python
# filter the data "AREA NAME", "AREA", "Crm Cd Desc"
filtered_df = df[["AREA NAME", "AREA", "Crm Cd Desc"]]
filtered_df.head()
```

```python
# group the data according to "AREA NAME"
grouped_by_area_df = filtered_df.groupby(["AREA NAME"]).size().reset_index(name="counts")
grouped_by_area_df.head()
```

```python
# bar chart
plt.figure(figsize=(20,10))
plt.bar(grouped_by_area_df["AREA NAME"], grouped_by_area_df["counts"])
plt.xticks(rotation=90)
plt.show()
```

```python
# pie chart
plt.figure(figsize=(20,10))
plt.pie(grouped_by_area_df["counts"], labels=grouped_by_area_df["AREA NAME"], autopct='%1.1f%%')
plt.show()
```

```python
# Plot
plt.figure(figsize=(20,10))
plt.plot(grouped_by_area_df["AREA NAME"], grouped_by_area_df["counts"])
plt.xticks(rotation=90)
plt.show()
```
