```python
# Hypothese: Men are more prone to get exposed to simple assault than women.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/LA_crime_data_2023.csv")
df.columns
```

```python
# "Vict Sex" and "Vict Descent" columns in df which is `Crm Cd Desc` =  `BATTERY - SIMPLE ASSAULT` on df
filtered_df = df[["Vict Sex", "Crm Cd Desc"]]
filtered_df = filtered_df[filtered_df["Crm Cd Desc"] == "BATTERY - SIMPLE ASSAULT"]
filtered_df.head()
```

```python
# group "Vict Sex" and count
grouped_by_sex = filtered_df.groupby(["Vict Sex"]).size().reset_index(name="counts")
grouped_by_sex.head()
```

```python
# bar chart
plt.figure(figsize=(10,5))
plt.bar(grouped_by_sex["Vict Sex"], grouped_by_sex["counts"])
plt.xticks(rotation=90)
plt.show()
```

```python
# pie chart
plt.figure(figsize=(20,10))
plt.pie(grouped_by_sex["counts"], labels=grouped_by_sex["Vict Sex"], autopct='%1.1f%%')
plt.show()
```
